```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('')
sales_filepath = Path('')

#  Initialize list objects to hold our menu and sales data
menu = []
sales = []

#  Read in the menu data into the menu list
with open('menu_data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    csv_header=next(csv_reader)
        
    for row in csv_reader:
        menu.append(row)

# Read in the sales data into the sales list
with open('sales_data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    csv_header = next(csv_reader)
        
    for row in csv_reader:
        sales.append(row)

#  Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize  variables
row_count = 0


# Loop over every row in the sales list object
for sales_row in sales:
    profit = 0
    cogs=0
    revenue=0
    item_found=False
    
    menu_item = sales_row[4]
    quantity = int(sales_row[3])



    # Loop over every row in the menu list object
    for menu_row in menu:
        item = menu_row[0]
        price = float(menu_row[3])
        cost = float(menu_row[4])
        
        # check if sales item matches menu item
        if item == menu_item:
            item_found=True
            profit += (price-cost) * quantity
            cogs += cost * quantity
            revenue += price * quantity
        

   
    if sales_row[4] not in report.keys():
        report[menu_item] = {
                "01-count": quantity,
                "02-revenue":revenue,
                "03-cogs": cogs,
                "04-profit": profit
            }
    else:
        report[menu_item]["01-count"]= report[menu_item]["01-count"] + quantity 
        report[menu_item]["02-revenue"]= report[menu_item]["02-revenue"] + revenue 
        report[menu_item]["03-cogs"]= report[menu_item]["03-cogs"] + cogs 
        report[menu_item]["04-profit"]= report[menu_item]["04-profit"] + profit 
        
    if not item_found:
        print(f"Sales item ->{menu_item} not found")
        
for key,value in report.items():
    print(key,value)
```
